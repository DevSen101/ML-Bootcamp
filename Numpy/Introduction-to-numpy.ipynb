### mean

```python
np.mean(a2)
```

### min
Return the minimum of an array or minimum along an axis.


```python
np.min(a2)
```

### max
Return the maximum of an array or maximum along an axis.

### std (Standard deviation)- Compute the standard deviation along the specified axis.
- how much the values in a dataset vary from the mean.
- a measure of how spread out a group of numbers is from the mean.

```python
np.std(a2)
```

### var - Variance
- Compute the variance along the specified axis.
- measure of the average degree to which each number is different to the mean.
- High variance = wider range of numbers
- Lower variance = lower range of numbers

```python
np.var(a2)
```

### Standard deviation = Squareroot of variance

```python
np.sqrt(np.var(a2))
```
