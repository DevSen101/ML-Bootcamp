# Matplotlib
- Matplotlib is a python data visualization library can create basic or advance plots (figures). which help us to visualize our data by converting numerical data into plot and visuals.
- Built on NumPy arrays (and Python)
- Integrates directly with pandas
- Simple to use interface(Once we get the foundation)

### Importing matplotlib, numpy and pandas

```python
# iumport matplotlib
%matplotlib inline # to matplotlib plots appear in our jupyter notebook enviroment
import matplotlib.pyplot as plt
# import numpy
import numpy as np
# import pandas
import pandas as pd
```

## First Method
### figure() 
- Create a new figure, or activate an existing figure.
### add_subplot()
- Add an axes to the figure as part of a subplot arrangement. 

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

## Second Method

### add_axes()
- Add an axes to the figure.
### plot()
- Add some data
- Plot y versus x as lines and/or markers.

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

## Third Method (Recommended)
### subplots()
- Create a figure and a set of subplots.
- This utility wrapper makes it convenient to create common layouts of subplots, including the enclosing figure object, in a single call.

```python

fig, ax = plt.subplots()
ax.plot(x, y); # add some data
```

## Matplotlib Example Workflow



```python
# Step -0 Import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# Step -1 Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Step -2 Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #(width, height)

# Step -3 Plot data
ax.plot(x, y)

# Step -4 Customize plot
ax.set(title="Sample Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# Step -5 Save & Show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

```python
import numpy as np
```

### linspace()
- Return evenly spaced numbers over a specified interval.

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

## Line Plot

```python
#  Plot the data and create the line plot.
fix, ax = plt.subplots()
ax.plot(x, x**2);
```

----------------------------------

## Scatter Plot

### scatter()
- A scatter plot of *y* vs. *x* with varying marker size and/or color.

```python
# Use same data to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fix, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

----------------------------------------------------

## Bar Plot

### bar()
- Make a bar plot.

```python
nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Store",
       xlabel="Items", ylabel="Price ($)");
```

### Same data with Horizontal representation.

### barh()
- Make a horizontal bar plot.

```python
fix, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values());
```

------------------------------------------


## Histogram

### random()
- Random Number Generation
### randn()
- Return a sample (or samples) from the "standard normal" distribution.
### hist()
- Compute and plot a histogram.

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Subplots Option 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10,15))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

### Subplots Option 2

```python
fig, ax = plt.subplots(2,2, figsize=(10,15))
# Plot to each different index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random,random(10), np.random.random(10));
```
