- Built on NumPy arrays (and Python)
- Integrates directly with pandas
- Simple to use interface(Once we get the foundation)

## First Method
### figure() 
- Create a new figure, or activate an existing figure.
### add_subplot()
- Add an axes to the figure as part of a subplot arrangement. 

```python
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

## Second Method

### add_axes()
- Add an axes to the figure.
### plot()
- Add some data
- Plot y versus x as lines and/or markers.

```python
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

## Third Method (Recommended)
### subplots()
- Create a figure and a set of subplots.
- This utility wrapper makes it convenient to create common layouts of subplots, including the enclosing figure object, in a single call.

```python

fig, ax = plt.subplots()
ax.plot(x, y); # add some data
```

## Matplotlib Example Workflow



```python
# Step -0 Import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# Step -1 Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# Step -2 Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) #(width, height)

# Step -3 Plot data
ax.plot(x, y)

# Step -4 Customize plot
ax.set(title="Sample Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# Step -5 Save & Show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

## Bar Plot

### bar()
- Make a bar plot.

```python
nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Store",
       xlabel="Items", ylabel="Price ($)");
```

```python
fix, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values());
```

### Which one should we use ? (pyplot Vs matplotlib OO method ?)
- When Plotting something quickly, okay to use pyplot method.
- When Plotting something more advanced, use the OO method.


```python
heart_desease
```

```python
over_50 = heart_desease[heart_desease["age"] > 50]
over_50.head()
```

### Pyplot method

```python
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

---
### OO method mixed with Pyplot method

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);

# ax.set_xlim([45, 100]); # Set the x-axis view limits.
```

### OO method from scratch

### axhline()
- Add a horizontal line spanning the whole or fraction of the Axes.
### set()
- Set multiple properties at once.

```python
fig, ax = plt.subplots(figsize=(10, 6))
# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c = over_50["target"],
                    )
# Customize the Plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a Horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

### Subplot of chol, age, thalach

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10, 10),
                              sharex=True) #sharex - share the x ax params.

# Add Data to ax1
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
       ylabel="Cholesterol")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])
# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(), linestyle="--");

# Add a title to the figure.
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```
